```python
train_data = pd.read_csv('/kaggle/input/telecon-data/train.csv')
```

```python
#Print the data to see if it loaded
print("This is all the data")
display(train_data)

# Print the data head to see the first 5 rows
print("This is data head")
display(train_data.head())

#Print out some basic stats about this data set
print("This is the data description")
display(train_data.describe())

#This prints out the names of the columns
print("Column names = ")
display(train_data.columns)
```

```python
train_data.drop(columns=['ID', 'customerID'], inplace = True, axis=1)
display(train_data.head())
```

```python
#Perform one-hot encoding for categorical columns
columns = ['gender', 'Partner', 'Dependents','PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod']
one_hot_1 = pd.get_dummies(train_data[['gender', 'Partner', 'Dependents','PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod']])

#Add the one-hot encoding categorical columns to the train data and remove the categorical columns
df = train_data.join(one_hot_1).drop(columns = columns)

#Display the improved data and 
display(df)
display(df.columns)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce', downcast='integer')
for col in df.columns:
    display(df[col][df[col].isnull()])
```

```python
df = df.dropna()

for col in df.columns:
    display(df[col][df[col].isnull()])
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()


continuous = ['tenure', 'MonthlyCharges', 'TotalCharges']
df[continuous] = min_max_scaler.fit_transform(df[continuous])

#This was to ensure that my code worked the way I wanted it to
display(df[continuous])

#I took my transformed continous variables and joined it with the one-hot encoding variables 
X_data = one_hot_1.join(df[continuous])


#I created my y_target to only include the income column which is my target variable
y_target = df['Churn']

#Used display to ensure my columns were printing the correct information
display(X_data)
display(y_target)
```

```python
for col in X_data.columns:
    display(X_data[col][X_data[col].isnull()])
```

```python
X_data = X_data.dropna()

for col in X_data.columns:
    display(X_data[col][X_data[col].isnull()])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_data, y_target, test_size=0.3, random_state=1, stratify=y_target)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

logreg = LogisticRegression(solver='lbfgs', multi_class='ovr')
logreg.fit(X_train, y_train)
logreg_predictions = logreg.predict(X_test)
accuracy = accuracy_score(y_test, logreg_predictions)
auc_logreg = roc_auc_score(y_test, logreg_predictions)

print(f"Logistic Regression Accuracy: {accuracy}")
print('Logistic Regression Misclassified examples: %d' % (y_test != logreg_predictions).sum())
print(f"Logistic Regression AUC: {auc_logreg}")
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score 

svm_classifier = SVC(random_state=1)
svm_classifier.fit(X_train, y_train)
svm_predictions = svm_classifier.predict(X_test)
accuracy = accuracy_score(y_test, svm_predictions)
auc = roc_auc_score(y_test, svm_predictions)

print(f"SVM Accuracy: {accuracy}")
print('SVM Misclassified examples: %d' % (y_test != svm_predictions).sum())
print(f"SVM AUC: {auc}")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

decision_tree = DecisionTreeClassifier(random_state=1)
decision_tree.fit(X_train, y_train)
decision_tree_predictions = decision_tree.predict(X_test)
accuracy = accuracy_score(y_test, decision_tree_predictions)
auc = roc_auc_score(y_test, decision_tree_predictions)

print(f"Decision Tree Accuracy: {accuracy}")
print('Decision Tree Misclassified examples: %d' % (y_test != decision_tree_predictions).sum())
print(f"Decision Tree AUC: {auc}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

knn_classifier = KNeighborsClassifier(n_neighbors=6)
knn_classifier.fit(X_train, y_train)
knn_predictions = knn_classifier.predict(X_test)
accuracy = accuracy_score(y_test, knn_predictions)
auc = roc_auc_score(y_test, knn_predictions)

print(f"KNN Accuracy: {accuracy}")
print('KKN Misclassified examples: %d' % (y_test != knn_predictions).sum())
print(f"KKN AUC: {auc}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

random_forest = RandomForestClassifier(random_state=1)
random_forest.fit(X_train, y_train)
random_forest_predictions = random_forest.predict(X_test)
accuracy = accuracy_score(y_test, random_forest_predictions)
auc = roc_auc_score(y_test, random_forest_predictions)

print(f"Random Forest Accuracy: {accuracy}")
print('KKN Misclassified examples: %d' % (y_test != random_forest_predictions).sum())
print(f"KKN AUC: {auc}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

pipe_lr = make_pipeline(StandardScaler(), LogisticRegression(random_state=1))

param_range = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

param_grid = {
    'logisticregression__C': param_range,
    'logisticregression__penalty': ['l1', 'l2']
}

gs = GridSearchCV(estimator=pipe_lr,
                  param_grid=param_grid,
                  scoring='roc_auc',  
                  cv=10)

gs = gs.fit(X_train, y_train)

print("Best ROC AUC Score:", gs.best_score_)
print("Best Hyperparameters:", gs.best_params_)

clf = gs.best_estimator_
print(f'Test accuracy: {clf.score(X_test, y_test):.3f}')
```
